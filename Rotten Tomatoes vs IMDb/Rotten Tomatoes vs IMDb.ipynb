# Rotten Tomatoes vs IMDb

I wanted to compare the ratings of movies between two popular websites which have slightly different agendas, Rotten tomatoes and Imdb. 

Rotten tomatoes is a 'Review aggregation website' whose primary function is to rate movies and tv shows. Here professional critics review and rate the content, and the website produces an aggregation of that score which they call the tomatometer. Rotten tomatoes also presents another score, the audience score, which is an aggregate of audience ratings. Rotten tomatoes presents information on cast, personal biographies, plot summaries, ratings of critics and fans. 

Imdb on the other hand is an online movie and tv show database (owned by Amazon). It has information on cast, production crew and personal biographies, plot summaries, trivia, ratings, and fan and critical reviews. Where as Rotten tomatoes is centered around critical reviews of a piece, Imdb is centered around being a database of everyone involved with the production of that piece. 

As you can see there is much overlap between the functions and features of the two websites. I had this impression that Rotten Tomatoes was more critical that Imdb, so I acquired two databases scraped from these two websites from kaggle and tried to answer the question. In this project I have explored just movies and not tv shows.

## Which website is more critcal

How do you determine if one website is more critical than another?

1) First I will find all the common movies between the two databases. 

2) I will take the difference between the tomatometer score and the imdb score, called diff. 

If the diff is positive for a particular movie it means that Rotten tomatoes (RT) score was higher then the Imdb score i.e. Imdb is more critical. If the diff was negative, then it means that Rotten tomatoes is more critical.

3) I will make a bar graph of the positive and negitive diff's to clearly see who is critical overall

4) I will order and plot this diff score as line graph and highlight the area under the curve

```python
# imdb data
imdb = pd.read_csv(r'C:\Users\harpo\Documents\Datasets\Movies\IMDb movies.csv', parse_dates=['year', 'date_published'])

# rotten tomatoes data
rt = pd.read_csv(r'C:\Users\harpo\Documents\Datasets\Movies\rotten_tomatoes_movies.csv', parse_dates=['in_theaters_date'])
```

```python
#display(imdb.head())
#display(rt.head())
```

```python
# rt ratings
rt_ratings = rt[['movie_title', 'tomatometer_rating', 'audience_rating', 'in_theaters_date', 'genre']]
rt_ratings.dropna(inplace=True)
rt_ratings['in_theaters_date'] = rt_ratings['in_theaters_date'].dt.year.astype(int)
rt_ratings = rt_ratings.rename(columns={'audience_rating':'rt_audience_rating', 'in_theaters_date':'year'})
rt_ratings['rt_audience_rating'] = rt_ratings['rt_audience_rating'].astype(int)
display(rt_ratings)
display(rt_ratings.info())

# imdb ratings
imdb_ratings = imdb[['title','avg_vote', 'year', 'genre']]
imdb_ratings = imdb_ratings.rename(columns={'title':'movie_title', 'avg_vote':'imdb_rating'})
imdb_ratings['year'] = imdb_ratings['year'].dt.year
display(imdb_ratings.info())
```

```python
rt_imdb_rate = pd.merge(rt_ratings, imdb_ratings, on=['movie_title', 'year'], suffixes=['_rt', '_imdb'])

display(rt_imdb_rate.count())
rt_imdb_rate
```

```python
# multiplied the imdb_rating by 10 because the rt score thier movies out of 100
# imdb score thier movies out of 10
rt_imdb_rate['imdb_rating'] = rt_imdb_rate['imdb_rating']*10
rt_imdb_rate['imdb_rating'] = rt_imdb_rate['imdb_rating'].astype(int)
rt_imdb_rate
```

Think of the index in the following plots as the movie ID. 

```python
# create a new column that is the difference between 
rt_imdb_rate['diff'] = rt_imdb_rate['tomatometer_rating'] - rt_imdb_rate['imdb_rating']
rt_imdb_rate.sort_values(by='diff', inplace=True)
rt_imdb_rate.reset_index(drop=True, inplace=True)
rt_imdb_rate
```

If I take a count of the movies that are positive and negitive I should get an idea of which website is more critical. We will visualise the data in a stacked bar chart. 

when:
diff -> positive: The tomatometer rated the movie higher than the imdb (imdb more critical)
diff -> negative: Imdb rated the movie higher than rt (rt more critical)



```python
from collections import OrderedDict

def calc_perc(df, df2):
    """
    Takes in a dataframe and calculates the percentage of that dataframe compared to a larger dataframe
    Parameters:
    df - a dataframe
    df2 - a dataframe
    
    Output:
    the ratio of length between df and df2, mulpulied by 100.   
    """
    return (len(df) / len(df2)) * 100
    

def split_df_in3(df):
    '''
    splits the dataframe into 3 smaller dataframes, where the 'diff' values of the dataframe are negative,
    positive and equal
    
    input:
    df - dataframe to be split
    
    output:
    a - dataframe where 'diff' < 0
    b - dataframe where 'diff' > 0
    c - dataframe where 'diff' == 0
    '''
    a = df[df['diff'] < 0]
    b = df[df['diff'] > 0]
    c = df[df['diff'] == 0]
    return a, b, c

# negitive diff -> RT more critical (RT_crit)
# positive diff -> Imdb more critical (Imdb_crit)
# diff == 0 -> the two websites agree (agreed)

split_df_lst = split_df_in3(rt_imdb_rate)

RT_crit, Imdb_crit, agreed = split_df_lst
    
# getting the percent of the dataframe rt_imdb_rate where 'diff', is negitive, positive or 0
RT_crit_score, Imdb_crit_score, agreed_score = [calc_perc(df, rt_imdb_rate) for df in split_df_lst]


df_dict = OrderedDict([(RT_crit_score, "RT's more Critical"), 
                      (Imdb_crit_score, "Imdb's more Critical"), 
                      (agreed_score, "Agreed")])

def stacked_bar(dic):
    a, b, c = dic.keys()
    diff_df = pd.DataFrame({f"{a:.2f}%: " + dic[a] : [a], 
                        f"{b:.2f}%: " + dic[b]: [b] ,
                        f"{c:.2f}%: " + dic[c] : [c]}, 
                      index=[''])

    diff_df.plot(kind='barh', stacked=True)
    sns.despine(left=True, bottom=True)
    plt.show()


stacked_bar(df_dict)
```

From this visualisation, you can see that rotton tomatoes is more critcal of movies than imdb in general. Though I personally thought that the difference would have been greater. The amount of perfect agreement between the two websites is very small.

I want to now plot the ordered dataframe rt_imdb_rate as a line graph and shade in the area under the curve. The area under the curve signifies how much more critical one site is compared to the other. 

```python
# I reset the index because I needed to cleanly plot the diff against the index (or movie ID)

def line_and_area_graph(df):
    dat = df.reset_index()
    sns.relplot(data=dat,
            x='index',
            y='diff',
            kind='line',
            height=11)
    plt.fill_between(dat['index'], dat['diff'], color='skyblue')
    plt.xlabel('Movie ID', size=14)
    plt.ylabel('diff', size=14)
    plt.show()
    
line_and_area_graph(rt_imdb_rate)
```

We can see the area under the curve of diff per Movie ID is greater for RT (area below the x-axis) compared to Imdb (area above the x-axis). We can see that when RT is more critical than Imdb, it is far more critical than Imdb, for a particular subset that is. This figure poses the question, are there particular subsets of movie genres that each site favour?

# Bias in criticism

Earlier I seperated the dataframe rt_imdb_rate into three smaller dataframes:
1) where diff was negitive (RT_crit)

2) where diff was positive (Imdb_crit)

3) where diff == 0 (agreed) (which we will ignore for this section as it was less then 2% of the data)

Now I am going to take these three dataframes and check if there is any biased behaviour in what the two sites rate the same movies. I will test if there is a bias the scores produced by each website depending on the movie genres. We should be able to see what movie genres appear more often in each dataframe

```python
# I am going to only use the RT labels, as they seem thorough enough
# I am going to ommit  the agreed dataframe for simplicity

df_lst = [RT_crit, Imdb_crit, rt_imdb_rate]

for df in df_lst:
    df['genre_rt'] = df['genre_rt'].str.split(', ')
    df.drop(columns=['genre_imdb'], inplace=True)
    
RT_crit
```

```python
# Now I want to count how many times each genre comes up
def genre_dict(df):
    '''
    Counts the number of each genre listed in the entire 'genre_rt' column. 
    
    Input:
    df - the dataframe which has a column 'genre_rt' which contains lists of genres.
    
    Output:
    genre_d - a dictionary of genres as keys and the count of how many times that genre appears in that 
                dataframe as the values.
    '''
    genre_d = {}

    for row in df['genre_rt']:
        for genre in row:
            if genre in genre_d.keys():
                genre_d[genre] += 1
            else:
                genre_d.update({genre: 1})
    return genre_d

# applying the genre dictionary function to the three data frames, not including the agreed dataframe, 
# and instead using the full dataframe rt_imdb_rate
RT_crit_genre_dict = genre_dict(RT_crit)
Imdb_crit_genre_dict = genre_dict(Imdb_crit)
full_genre_dict = genre_dict(rt_imdb_rate)

# creating dataframes out of the dictionaries
RT_crit_genre_df = pd.DataFrame(RT_crit_genre_dict.items(), columns=['Genres', 'RT genre count'])
Imdb_crit_genre_df = pd.DataFrame(Imdb_crit_genre_dict.items(), columns=['Genres', 'Imdb genre count'])
full_genre_df = pd.DataFrame(full_genre_dict.items(), columns=['Genres', 'Count'])

# Merging the RT_crit_genre_df and the Imdb_cirt_genre_df, and setting the genres as the index
genre_df = pd.merge(RT_crit_genre_df, Imdb_crit_genre_df, on='Genres', how='outer')
genre_df.set_index('Genres', inplace=True)

# Setting the Genres column as the index for the full_genre_df
full_genre_df.set_index('Genres', inplace=True)
```

For context I plotted the popularity of movie genres.

```python
# sort values by Count
full_genre_df.sort_values(by='Count', inplace=True)

full_genre_df.plot(kind='barh', legend=False)
plt.xlabel('Count', size=13)
plt.ylabel('Genres', size=13)
plt.show()

#full_genre_df
```

```python
# the final counts of how many times each genre shows up in each dataframe
display(genre_df)
```

```python
# creating a df for a filled barchart as we are only interested in proportions
filled_stacked_genre_df = genre_df.apply(lambda x: x*100/sum(x), axis=1)

# plotting the filled barchart
filled_stacked_genre_df.plot(kind='barh', stacked=True)
plt.legend(loc='lower right')
plt.show()
```

From this graph we can see some small differences, where one site is more critical than another site. 
Imdb is almost disproportionally more critical on one particular genre: Classics. To a lesser extent Imdb is also more critical on Westerns. RT on the other hand is more ciritcal on Television and Gay & Lesbian. 

From this high level glance I would really only be confident in saying that Imdb is more critical of the classics genre. All other genres I can possibly say that one site may be slightly more critical that another, and that is not taking into account that there are many more variables that contribute to the popularity of a movie.
