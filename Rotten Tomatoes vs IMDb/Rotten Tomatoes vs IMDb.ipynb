# Rotten Tomatoes vs IMDb

I wanted to compare the ratings of movies between two popular websites which have slightly different agendas, Rotten tomatoes and Imdb. 

Rotten tomatoes is a 'Review aggregation website' whose primary function is to rate movies and tv shows. Here professional critics review and rate the content, and the website produces an aggregation of that score which they call the tomatometer. Rotten tomatoes also presents another score, the audience score, which is an aggregate of audience ratings. Rotten tomatoes presents information on cast, personal biographies, plot summaries, ratings of critics and fans. 

Imdb on the other hand is an online movie and tv show database (owned by Amazon). It has information on cast, production crew and personal biographies, plot summaries, trivia, ratings, and fan and critical reviews. Where as Rotten tomatoes is centered around critical reviews of a piece, Imdb is centered around being a database of everyone involved with the production of that piece. 

As you can see there is much overlap between the functions and features of the two websites. I had this impression that Rotten Tomatoes was more critical that Imdb, so I acquired two databases scraped from these two websites from kaggle and tried to answer the question. In this project I have explored just movies and not tv shows.

## Which website is more critcal

How do you determine if one website is more critical than another?

1) First I will find all the common movies between the two databases. 

2) I will take the difference between the tomatometer score and the imdb score, called diff. 

If the diff is positive for a particular movie it means that Rotten tomatoes (RT) score was higher then the Imdb score i.e. Imdb is more critical. If the diff was negative, then it means that Rotten tomatoes is more critical.

3) I will make a bar graph of the positive and negitive diff's to clearly see who is critical overall

4) I will order and plot this diff score as line graph and highlight the area under the curve

```python
# imdb data
imdb = pd.read_csv(r'C:\Users\harpo\Documents\Datasets\Movies\IMDb movies.csv', parse_dates=['year', 'date_published'])

# rotten tomatoes data
rt = pd.read_csv(r'C:\Users\harpo\Documents\Datasets\Movies\rotten_tomatoes_movies.csv', parse_dates=['in_theaters_date'])
```

```python
#display(imdb.head())
#display(rt.head())
```

```python
# rt ratings
rt_ratings = rt[['movie_title', 'tomatometer_rating', 'audience_rating', 'in_theaters_date', 'genre']]
rt_ratings.dropna(inplace=True)
rt_ratings['in_theaters_date'] = rt_ratings['in_theaters_date'].dt.year.astype(int)
rt_ratings = rt_ratings.rename(columns={'audience_rating':'rt_audience_rating', 'in_theaters_date':'year'})
display(rt_ratings)
display(rt_ratings.info())

# imdb ratings
imdb_ratings = imdb[['title','avg_vote', 'year', 'genre']]
imdb_ratings = imdb_ratings.rename(columns={'title':'movie_title', 'avg_vote':'imdb_rating'})
imdb_ratings['year'] = imdb_ratings['year'].dt.year
display(imdb_ratings.info())
```

```python
rt_imdb_rate = pd.merge(rt_ratings, imdb_ratings, on=['movie_title', 'year'], suffixes=['_rt', '_imdb'])

display(rt_imdb_rate.count())
rt_imdb_rate
```

```python
# multiplied the imdb_rating by 10 because the rt score thier movies out of 100
# imdb score thier movies out of 10
rt_imdb_rate['imdb_rating'] = rt_imdb_rate['imdb_rating']*10
rt_imdb_rate[['rt_audience_rating', 'imdb_rating']] = rt_imdb_rate[['rt_audience_rating', 'imdb_rating']].astype(int)
rt_imdb_rate
```

Think of the index in the following plots as the movie ID. 

```python
# create a new column that is the difference between 
rt_imdb_rate['diff'] = rt_imdb_rate['tomatometer_rating'] - rt_imdb_rate['imdb_rating']
rt_imdb_rate.sort_values(by='diff', inplace=True)
rt_imdb_rate.reset_index(drop=True, inplace=True)
rt_imdb_rate
```

If I take a count of the movies that are positive and negitive I should get an idea of which website is more critical. We will visualise the data in a stacked bar chart. 

when:
diff -> positive: The tomatometer rated the movie higher than the imdb (imdb more critical)
diff -> negative: Imdb rated the movie higher than rt (rt more critical)



```python
sns.set_style('white')

def calc_perc(df, df2=rt_imdb_rate):
    """
    Takes in a dataframe and calculates the percentage of that dataframe compared to a larger dataframe
    Parameters:
    df - a dataframe
    df2 - a dataframe
    
    Output:
    the ratio of length between df and df2, mulpulied by 100.   
    """
    return (len(df) / len(df2)) * 100
    

# negitive diff -> RT more critical 
RT_crit = rt_imdb_rate[rt_imdb_rate['diff'] > 0]
RT_crit_score = calc_perc(RT_crit)

# positive diff -> Imdb more critical
Imdb_crit = rt_imdb_rate[rt_imdb_rate['diff'] < 0]
Imdb_crit_score = calc_perc(Imdb_crit)

# diff == 0 -> the two websites agree
agreed = rt_imdb_rate[rt_imdb_rate['diff'] == 0]
agreed_score = calc_perc(agreed)


diff_df = pd.DataFrame({"{:.2f}%: Imdb's more critical".format(Imdb_crit_score) : [Imdb_crit_score], 
                        "{:.2f}%: RT's more Critical".format(RT_crit_score):[RT_crit_score] ,
                        "{:.2f}%: Agreed".format(agreed_score) : [agreed_score]}, 
                      index=[''])

diff_df.plot(kind='barh', stacked=True)
sns.despine(left=True, bottom=True)
plt.show()
```

From this visualisation, you can see that rotton tomatoes is more critcal of movies than imdb in general. Though I personally thought that the difference would have been greater. The amount of perfect agreement between the two websites is very small.

I want to now plot the ordered dataframe rt_imdb_rate as a line graph and shade in the area under the curve. The area under the curve signifies how much more critical one site is compared to the other. 

```python
# I reset the index because I needed to cleanly plot the diff against the index (or movie ID)
dat = rt_imdb_rate.reset_index()
sns.relplot(data=dat,
            x='index',
            y='diff',
            kind='line',
            height=11)
plt.fill_between(dat['index'], dat['diff'], color='skyblue')
plt.xlabel('Movie ID', size=14)
plt.ylabel('diff', size=14)
plt.show()
```

We can see the area under the curve of diff per Movie ID is greater for RT (area below the x-axis) compared to Imdb (area above the x-axis). We can see that when RT is more critical than Imdb, it is far more critical than Imdb, for a particular subset that is. This figure poses the question, are there particular subsets of movie genres that each site favour?
