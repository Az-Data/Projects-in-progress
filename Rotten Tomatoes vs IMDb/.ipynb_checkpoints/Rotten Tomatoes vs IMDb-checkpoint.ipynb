# Rotten Tomatoes vs IMDb

I wanted to compare the ratings of movies between two popular websites which have slightly different agendas, Rotten tomatoes and Imdb. 

Rotten tomatoes is a 'Review aggregation website' whose primary function is to rate movies and tv shows. Here professional critics review and rate the content, and the website produces an aggregation of that score which they call the tomatometer. Rotten tomatoes also presents another score, the audience score, which is an aggregate of audience ratings. Rotten tomatoes presents information on cast, personal biographies, plot summaries, ratings of critics and fans. 

Imdb on the other hand is an online movie and tv show database (owned by Amazon). It has information on cast, production crew, personal biographies, plot summaries, trivia, ratings, as well as fan and critical reviews. Where as Rotten tomatoes is centered around critical reviews of a piece, Imdb is centered around being a database of everyone involved with the production of that piece. 

As you can see there is much overlap between the functions and features of the two websites. I had this impression that Rotten Tomatoes was more critical than Imdb, so I acquired two datasets scraped from these two websites (sourced from kaggle) and tried to answer the question. In this project I have explored just movies and not tv shows.

## Which website is more critcal

How do you determine if one website is more critical than another?

1) First I will find all the common movies between the two databases. 

2) I will take the difference between the tomatometer score and the imdb score, called diff. 

If the diff is positive for a particular movie it means that Rotten tomatoes (RT) score was higher then the Imdb score i.e. Imdb was more critical on that movie. If the diff was negative, then it means that Rotten tomatoes was more critical on that movie.

3) I will make a bar graph of the positive and negitive diff's to clearly see who is more critical overall

4) I will order and plot this diff score as a line graph and fill the area under the curve
This will give an indication of the spread of the difference (diff) between the scores for all the movies.

### Population statistics and z-score preparation

We are going to take the population mean of each of the datasets to calculate the z-score later. Here we assume that the full dataset is the population mean. 

The z-score is defined by:

$$z=\frac{x-\mu}{\sigma}$$

Here $\mu$ is the population mean, $\sigma$ is the population standard deviation, and $x$ is the sample size.

Here is the tricky bit. Since the two datasets are very different sizes. Imdb being much bigger than rt dataset. I am going to bootstrap and random select a sample of the Imdb dataset to match the size of the rt dataset, before preforming the mean and standard deviation calculations.

```python
def mean_reduced_population(data_series, sample_size):
    """
    Calculates the mean by taking a sample of the dataset and running it 1000 times
    """
    mean_list = [data_series.sample(n=sample_size).mean() for a in range(1000)]
    return np.mean(mean_list)

def std_reduced_population(data_series, sample_size):
    """
    Calculates the std by taking a sample of the dataset and running it 1000 times
    """
    mean_list = [data_series.sample(n=sample_size).std() for a in range(1000)]
    return np.mean(mean_list)
    
mu_imdb_bootstrapped = mean_reduced_population(imdb_ratings['imdb_rating']*10, len(rt_ratings))

sigma_imdb_bootstrapped = std_reduced_population(imdb_ratings['imdb_rating']*10, len(rt_ratings))



#population mean of imdb scores
mu_imdb_test = imdb_ratings['imdb_rating'].mean()*10

#population standard deviation for imdb
sigma_imdb_test = imdb_ratings['imdb_rating'].std()*10


#population mean of rotten tomatoes scores
mu_rt = rt_ratings['tomatometer_rating'].mean()

#populaiton standard deviation for rt
sigma_rt = rt_ratings['tomatometer_rating'].std()

print(mu_imdb_bootstrapped)
print(mu_imdb_test)
print(" ")
print(sigma_imdb_bootstrapped)
print(sigma_imdb_test)
```

### Comparing scores on common movies

```python
rt_imdb_rate = pd.merge(rt_ratings, imdb_ratings, on=['movie_title', 'year'], suffixes=['_rt', '_imdb'])

rt_imdb_rate
```

```python
# multiplied the imdb_rating by 10 to get it as the same format as the RT tomatometer score
rt_imdb_rate['imdb_rating'] = rt_imdb_rate['imdb_rating']*10
rt_imdb_rate['imdb_rating'] = rt_imdb_rate['imdb_rating'].astype(int)
rt_imdb_rate
```

Think of the index in the following plots as the movie ID. 

```python
# create a new column that is the difference between 
rt_imdb_rate['diff'] = rt_imdb_rate['tomatometer_rating'] - rt_imdb_rate['imdb_rating']
rt_imdb_rate.sort_values(by='diff', inplace=True)
rt_imdb_rate.reset_index(drop=True, inplace=True)
rt_imdb_rate
```

If I take a count of diff which are negative or positive we should get an idea of which website is more critical. We will visualise the data in a stacked bar chart. 

Just to clarify, when:

diff -> positive: The tomatometer rated the movie higher than the imdb (imdb more critical)

diff -> negative: Imdb rated the movie higher than rt (rt more critical)

```python
from collections import OrderedDict

def calc_perc(df, df2):
    """
    Takes in a dataframe and calculates the percentage of that dataframe compared to a larger dataframe
    
    Input:
    df - a dataframe
    df2 - a dataframe
    
    Output:
    the ratio of length between df and df2, mulpulied by 100.   
    """
    return (len(df) / len(df2)) * 100
    

def split_df_in3(df):
    '''
    splits the dataframe into 3 smaller dataframes, where the 'diff' values of the dataframe are negative,
    positive and equal
    
    input:
    df - dataframe to be split
    
    output:
    a - dataframe where 'diff' < 0
    b - dataframe where 'diff' > 0
    c - dataframe where 'diff' == 0
    '''
    a = df[df['diff'] < 0]
    b = df[df['diff'] > 0]
    c = df[df['diff'] == 0]
    return a, b, c

# negitive diff -> RT more critical (RT_crit)
# positive diff -> Imdb more critical (Imdb_crit)
# diff == 0 -> the two websites agree (agreed)

split_df_lst = split_df_in3(rt_imdb_rate)

RT_crit, Imdb_crit, agreed = split_df_lst
    
# getting the percent of the dataframe rt_imdb_rate where 'diff', is negitive, positive or 0
RT_crit_score, Imdb_crit_score, agreed_score = [calc_perc(df, rt_imdb_rate) for df in split_df_lst]


df_dict = OrderedDict([(RT_crit_score, "RT's more Critical"), 
                      (Imdb_crit_score, "Imdb's more Critical"), 
                      (agreed_score, "Agreed")])

def stacked_bar(dic):
    '''
    Takes in a dictionary of proportions as keys and a description as values and produces a stacked barchart.
    
    Input:
    dic - Dictionary with key = score, and value = string
    
    Ouptut:
    Plots a stacked barchart
    '''
    a, b, c = dic.keys()
    diff_df = pd.DataFrame({f"{a:.2f}%: " + dic[a] : [a], 
                        f"{b:.2f}%: " + dic[b]: [b] ,
                        f"{c:.2f}%: " + dic[c] : [c]}, 
                      index=[''])

    diff_df.plot(kind='barh', stacked=True)
    sns.despine(left=True, bottom=True)
    plt.show()


stacked_bar(df_dict)
```

There is also a statistical test we can do which will help us determine if the two data sets are statistically different. 

From this visualisation, you can see that rotton tomatoes is more critcal of movies than imdb in general. Though I personally thought that the difference would have been greater. The amount of perfect agreement between the two websites is very small.

I want to now plot the ordered dataframe rt_imdb_rate as a line graph and shade in the area under the curve. The area under the curve signifies how much more critical one site is compared to the other. The line graph will also give an indication of the spread of the data. 

```python
# I reset the index because I needed to cleanly plot the diff against the index (or movie ID)

def line_and_area_graph(df):
    '''
    Takes in a dataframe and plots it against its index, and fills in the area under the curve
    
    Input:
    df - dataframe to converted into a line and area graph, must have the column 'diff' and df should already be
            sorted for best results
            
    Output:
    Line graph with the area between the line and the x-axis filled in
    '''
    dat = df.reset_index()
    sns.relplot(data=dat,
            x='index',
            y='diff',
            kind='line',
            height=11)
    plt.fill_between(dat['index'], dat['diff'], color='skyblue')
    plt.xlabel('Movie ID', size=14)
    plt.ylabel('diff', size=14)
    sns.despine(bottom=True, left=True)
    plt.show()
    
line_and_area_graph(rt_imdb_rate)
```

We can see the area between the curve and the x-axis is greater for RT (area below the x-axis) compared to Imdb (area above the x-axis). It is also interesting to note how few movies are at the extremes, and that they are almost equal and opposite. 

A question that could be posed: are there particular subsets of movie genres that each site favour?

# Bias in criticism

Earlier I seperated the dataframe rt_imdb_rate into three smaller dataframes:
1) where diff was negitive (RT_crit)

2) where diff was positive (Imdb_crit)

3) where diff == 0 (agreed) (which we will ignore for this section as it was less then 2% of the data)

Now I am going to take these two dataframes and see if either site is more critical toward a particular genre of movie. I will make a frequency count of all the genres in the two dataframes. We should be able to see what movie genres appear more often in each dataframe.

```python
# I am going to only use the RT labels, as they seem thorough enough
# I am going to ommit the agreed dataframe for simplicity
# I also used to full dataframe rt_imdb_rate to plot all the frequencies of genres for context

df_lst = [RT_crit, Imdb_crit, rt_imdb_rate]

# producing a list of genres in the 'genre_rt' column
for df in df_lst:
    df['genre_rt'] = df['genre_rt'].str.split(', ')
    df.drop(columns=['genre_imdb'], inplace=True)
    
#RT_crit
```

```python
# Now I want to count how many times each genre comes up and put it in a dictionary
def genre_dict(df):
    '''
    Counts the number of each genre listed in the entire 'genre_rt' column. 
    
    Input:
    df - the dataframe which has a column 'genre_rt' which contains lists of genres.
    
    Output:
    genre_d - a dictionary of genres as keys and the count of how many times that genre appears in that 
                dataframe as the values.
    '''
    genre_d = {}

    for row in df['genre_rt']:
        for genre in row:
            if genre in genre_d.keys():
                genre_d[genre] += 1
            else:
                genre_d.update({genre: 1})
    return genre_d

# applying the genre dictionary function to the three dataframes, and producing dictionaries
RT_crit_genre_dict = genre_dict(RT_crit)
Imdb_crit_genre_dict = genre_dict(Imdb_crit)
full_genre_dict = genre_dict(rt_imdb_rate)

# creating dataframes out of the dictionaries
RT_crit_genre_df = pd.DataFrame(RT_crit_genre_dict.items(), columns=['Genres', 'RT genre count'])
Imdb_crit_genre_df = pd.DataFrame(Imdb_crit_genre_dict.items(), columns=['Genres', 'Imdb genre count'])
full_genre_df = pd.DataFrame(full_genre_dict.items(), columns=['Genres', 'Count'])

# Merging the RT_crit_genre_df and the Imdb_cirt_genre_df, and setting the genres as the index
genre_df = pd.merge(RT_crit_genre_df, Imdb_crit_genre_df, on='Genres', how='outer')
genre_df.set_index('Genres', inplace=True)

# Setting the Genres column as the index for the full_genre_df
full_genre_df.set_index('Genres', inplace=True)
```

For context I plotted the popularity of movie genres for the full dataframe rt_imdb_rate

```python
# sort values by Count
full_genre_df.sort_values(by='Count', inplace=True)

# Bar plot of genres 
full_genre_df.plot(kind='barh', legend=False)
plt.xlabel('Count', size=13)
plt.ylabel('Genres', size=13)
sns.despine(bottom=True, left=True)
plt.show()

#full_genre_df
```

It is interesting to note that genres Drama and Comedy come up the most.

```python
# creating a df for a filled barchart as we are only interested in proportions
filled_stacked_genre_df = genre_df.apply(lambda x: x*100/sum(x), axis=1)

# plotting the filled barchart
filled_stacked_genre_df.plot(kind='barh', stacked=True)
plt.legend(loc='lower right')
sns.despine(bottom=True, left=True)
plt.show()
```

From this graph we can see some differences in proportion, where one site is more critical than another site. However Imdb is almost disproportionally more critical on one particular genre: Classics. To a lesser extent Imdb is also more critical on Westerns. RT on the other hand is more ciritcal on the genres of Television and Gay & Lesbian. 

From this high level glance I would really only be confident in saying that Imdb is more critical of the classics genre. All other genres I can possibly say that one site may be slightly more critical that another, and that is not taking into account that there are many more variables that contribute to the popularity of a movie. ## Tomatometer vs audience rating

In this section I want to just look at the Rotten Tomatoes dataset. RT displays the Tomatometer score, produced by professional critics, and the Audience score, produced by passionate movie viewers. I would like to compare the two scores to see how they differ.

```python
rt_ratings
```

```python
rt_ratings['diff'] = rt_ratings['tomatometer_rating'] - rt_ratings['rt_audience_rating']
rt_ratings
```

```python
# negitive diff -> RT Audience more critical 
# positive diff -> RT critics more critical
# diff == 0 -> RT critics and audience agree

RT_df_lst = split_df_in3(rt_ratings)

RT_audience, RT_critics, critics_audience_agreed = RT_df_lst

# Finding the percentage of rt_ratings that 'diff' is negative, positive and equal to 0
RT_audience_score, RT_critics_score, critics_audience_agreed_score = [calc_perc(df=i, df2=rt_ratings)
                                                                         for i in RT_df_lst]

RT_crit_aud_dict = OrderedDict([
                                (RT_critics_score, "Critics are more critical"),
                                (RT_audience_score, "Audience are more critical"),
                                (critics_audience_agreed_score, "Critics and Audience agree")
                                ])

stacked_bar(RT_crit_aud_dict)
```

Not suprisingly the critics are more critical than the audience. Though the difference is not as stark as I would have thought it would be.
