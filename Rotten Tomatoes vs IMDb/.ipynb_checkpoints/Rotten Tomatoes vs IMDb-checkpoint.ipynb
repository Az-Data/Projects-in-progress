# Rotten Tomatoes vs IMDb

I wanted to compare the ratings of common movies between rotten tomatoes and Imdb and see which website is the most critical with thier ratings. 


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = [14, 10]
```

```python
# imdb data
imdb = pd.read_csv(r'C:\Users\harpo\Documents\Datasets\Movies\IMDb movies.csv', parse_dates=['year', 'date_published'])
imdb_short = imdb[['title', 'original_title', 'year', 'genre', 'country', 'language',
                  'avg_vote', 'votes', 'budget']]

# rotten tomatoes data
rt = pd.read_csv(r'C:\Users\harpo\Documents\Datasets\Movies\rotten_tomatoes_movies.csv', parse_dates=['in_theaters_date'])
rt_short = rt[['movie_title','genre','tomatometer_status', 'in_theaters_date',
       'tomatometer_rating', 'tomatometer_count', 'audience_status', 'audience_rating', 'audience_count']]
```

```python
#display(imdb.columns)
display(rt)
imdb_short.head()
```

```python
# rt ratings
rt_ratings = rt[['movie_title', 'tomatometer_rating', 'audience_rating', 'in_theaters_date']]
rt_ratings['in_theaters_date'] = rt_ratings['in_theaters_date'].dt.year.astype('int')
display(rt_ratings)
rt_ratings = rt_ratings.rename(columns={'audience_rating':'rt_audience_rating', 'in_theaters_date':'year'})
display(rt_ratings.info())

# imdb ratings
imdb_ratings = imdb[['title','avg_vote', 'year']]
imdb_ratings = imdb_ratings.rename(columns={'title':'movie_title', 'avg_vote':'imdb_rating'})
imdb_ratings['year'] = imdb_ratings['year'].dt.year
display(imdb_ratings.info())
```

```python
rt_imdb_rate = pd.merge(rt_ratings, imdb_ratings, on='movie_title')

display(rt_imdb_rate.count())
rt_imdb_rate
```
